### 計算詞頻 ###

```python
from nltk.tokenize import word_tokenize

file = open('overall.txt', 'r')
text = file.read()
#print(text)
chlist = word_tokenize(text)

def list2freqdict(mylist):
    mydict=dict()
    for ch in mylist:
        mydict[ch]=mydict.get(ch,0)+1
    return mydict

chfreqdict_original=list2freqdict(chlist)
print(chfreqdict_original)

#將frequency排序
from operator import itemgetter
chfreqsorted=sorted(chfreqdict_original.items(), key=itemgetter(1), reverse=True)
print(chfreqsorted)

#另存一rank之dicctionary
chfreqdict_rank = chfreqdict_original
count = 1

print ("================")
for k in chfreqsorted:
    #print (k[0])
    chfreqdict_rank[ k[0] ] = count
    count += 1

print (chfreqdict_rank)
```

### TXT_Encoding ###

```python
import os
for family in os.listdir(text_directory):
    directory = "C:/Users/Julie/Dropbox/Python/Tensor/txt/" 
    
    directory = directory + family + '/'
    print ("===================")
    print (family)
    print ("===================")
    
    hl_list = next(os.walk(directory))[2] # get all filenames in the in_directory
    hl_list = [os.path.join(directory, f) for f in hl_list] # filepathname list

    hl_list = list(filter(lambda f: f.endswith(".txt"), hl_list)) # in case some non-hooklog file in the folder
    #print (hl_list)
    
    for txt_file in hl_list:
        each_txt_file = open(txt_file, 'r')
        text = each_txt_file.read()
        print ("--------")
        #print("text_before：" + text)
        for k in chfreqdict_rank.keys():
            text = text.replace(k,str(chfreqdict_rank.get(k)))
        #print("text_after：" + test_text)
        print (text)
        
        en_directory = "C:/Users/Julie/Dropbox/Python/Tensor/Encoding/" #hooklog輸入
        
        en_directory = en_directory + family
        if not os.path.isdir(en_directory):
            os.mkdir(en_directory)
        #print (en_directory)
        
        en_directory = en_directory + '/' +txt_file.split("/")[8]
        file = open(en_directory, 'w', encoding = 'UTF-8')
        file.write(text)
        
        file.close()
        each_txt_file.close()
```
